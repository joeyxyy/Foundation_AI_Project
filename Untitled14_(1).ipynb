```python
# ============================================================================
# SECTION 2: DATA COLLECTION FROM USGS (WITH CACHING AND ERROR HANDLING)
# ============================================================================

import os
import time

def fetch_usgs_earthquake_data(start_date, end_date, min_lat, max_lat, min_lon, max_lon,
                               min_mag=1.5, save_csv=True, csv_path='earthquakes_bay_area.csv',
                               use_cache=True):
    """
    Fetch earthquake data from USGS API with intelligent caching and chunked fetching

    Parameters:
    -----------
    start_date, end_date : str
        Date range in YYYY-MM-DD format
    min_lat, max_lat, min_lon, max_lon : float
        Bounding box coordinates for the region
    min_mag : float
        Minimum magnitude threshold (default: 1.5 for catalog completeness)
    save_csv : bool
        Whether to save fetched data to CSV for reproducibility
    csv_path : str
        Path to save/load CSV file
    use_cache : bool
        If True, load from CSV if it exists (saves time during development)

    Returns:
    --------
    DataFrame with earthquake events

    Design Notes:
    - USGS API has 20,000 event limit - we fetch in yearly chunks
    - CSV caching enables reproducibility and faster development iteration
    - Handles API rate limiting and network errors gracefully
    """

    # Check if cached data exists
    if use_cache and os.path.exists(csv_path):
        print(f"Found cached data at: {csv_path}")
        print(f"Loading from cache (use use_cache=False to re-fetch)...")
        try:
            df = pd.read_csv(csv_path)
            df['time'] = pd.to_datetime(df['time'])
            print(f"✓ Loaded {len(df)} cached earthquake events")
            print(f"✓ Date range: {df['time'].min()} to {df['time'].max()}")
            return df
        except Exception as e:
            print(f"⚠ Error loading cache: {e}")
            print(f"Fetching fresh data from USGS API...")

    # Fetch from USGS API in yearly chunks (to avoid 20,000 limit)
    base_url = "https://earthquake.usgs.gov/fdsnws/event/1/query"

    print(f"\nFetching earthquake data from USGS API...")
    print(f"  Date range: {start_date} to {end_date}")
    print(f"  Region: Lat [{min_lat}, {max_lat}], Lon [{min_lon}, {max_lon}]")
    print(f"  Minimum magnitude: {min_mag}")
    print(f"  Note: USGS API has 20,000 event limit - fetching in yearly chunks...")

    # Parse dates
    start = pd.to_datetime(start_date)
    end = pd.to_datetime(end_date)

    all_events = []

    # Fetch year by year
    current_start = start
    year_count = 0

    while current_start < end:
        current_end = min(current_start + pd.DateOffset(years=1), end)
        year_count += 1

        params = {
            'format': 'geojson',
            'starttime': current_start.strftime('%Y-%m-%d'),
            'endtime': current_end.strftime('%Y-%m-%d'),
            'minlatitude': min_lat,
            'maxlatitude': max_lat,
            'minlongitude': min_lon,
            'maxlongitude': max_lon,
            'minmagnitude': min_mag,
            'orderby': 'time'
        }

        print(f"  Fetching year {current_start.year}... ", end='', flush=True)

        try:
            response = requests.get(base_url, params=params, timeout=120)

            if response.status_code == 200:
                data = response.json()
                year_events = len(data['features'])
                print(f"✓ {year_events} events")

                for feature in data['features']:
                    props = feature['properties']
                    coords = feature['geometry']['coordinates']

                    all_events.append({
                        'time': pd.to_datetime(props['time'], unit='ms'),
                        'latitude': coords[1],
                        'longitude': coords[0],
                        'depth': coords[2] if coords[2] is not None else 0.0,
                        'magnitude': props['mag'],
                        'place': props.get('place', 'Unknown'),
                        'type': props.get('type', 'earthquake')
                    })

            elif response.status_code == 400:
                print(f"✗ Error 400")
                print(f"  Response: {response.text[:300]}")
                print(f"  Skipping this year...")

            else:
                print(f"✗ HTTP {response.status_code}")

        except Exception as e:
            print(f"✗ Error: {e}")

        current_start = current_end
        time.sleep(0.5)  # Be nice to the API

    if len(all_events) == 0:
        print(f"✗ No events fetched")
        return None

    # Create DataFrame
    df = pd.DataFrame(all_events)

    # Remove any duplicate events
    initial_len = len(df)
    df = df.drop_duplicates(subset=['time', 'latitude', 'longitude', 'magnitude'])
    if len(df) < initial_len:
        print(f"  Removed {initial_len - len(df)} duplicate events")

    print(f"\n✓ Successfully fetched {len(df)} total earthquake events")

    # Save to CSV for future use
    if save_csv:
        try:
            df.to_csv(csv_path, index=False)
            print(f"✓ Saved data to: {csv_path}")
            print(f"  (Use use_cache=True to load from this file next time)")
        except Exception as e:
            print(f"⚠ Warning: Could not save CSV: {e}")

    return df

# San Francisco Bay Area bounding box
# Covers major faults: San Andreas, Hayward, Calaveras
MIN_LAT, MAX_LAT = 36.5, 38.5
MIN_LON, MAX_LON = -123.5, -121.0

print("\n" + "=" * 80)
print("DATA COLLECTION: USGS EARTHQUAKE CATALOG")
print("=" * 80 + "\n")

# Fetch data with intelligent caching
earthquakes_df = fetch_usgs_earthquake_data(
    start_date='2000-01-01',
    end_date='2024-11-30',
    min_lat=MIN_LAT,
    max_lat=MAX_LAT,
    min_lon=MIN_LON,
    max_lon=MAX_LON,
    min_mag=1.5,
    save_csv=True,
    csv_path='earthquakes_bay_area_2000_2024.csv',
    use_cache=True  # Set to False to force re-fetch from USGS
)

if earthquakes_df is None:
    raise RuntimeError("Failed to fetch earthquake data. Check error messages above.")

# Data validation and summary
print(f"\n" + "=" * 80)
print("DATA SUMMARY AND EXPORT")
print("=" * 80)
print(f"Dataset shape: {earthquakes_df.shape}")
print(f"Date range: {earthquakes_df['time'].min()} to {earthquakes_df['time'].max()}")
print(f"Duration: {(earthquakes_df['time'].max() - earthquakes_df['time'].min()).days / 365.25:.1f} years")
print(f"\nMagnitude distribution:")
print(earthquakes_df['magnitude'].describe())
print(f"\nDepth distribution (km):")
print(earthquakes_df['depth'].describe())
print(f"\nEvent types:")
print(earthquakes_df['type'].value_counts())

# Export summary statistics to CSV
summary_stats = pd.DataFrame({
    'Metric': ['Total Events', 'Date Range Start', 'Date Range End', 'Duration (years)',
               'Min Magnitude', 'Max Magnitude', 'Mean Magnitude', 'Median Magnitude',
               'M>=3.0 Events', 'M>=3.0 Percentage'],
    'Value': [
        len(earthquakes_df),
        earthquakes_df['time'].min(),
        earthquakes_df['time'].max(),
        f"{(earthquakes_df['time'].max() - earthquakes_df['time'].min()).days / 365.25:.1f}",
        earthquakes_df['magnitude'].min(),
        earthquakes_df['magnitude'].max(),
        f"{earthquakes_df['magnitude'].mean():.2f}",
        f"{earthquakes_df['magnitude'].median():.2f}",
        (earthquakes_df['magnitude'] >= 3.0).sum(),
        f"{100 * (earthquakes_df['magnitude'] >= 3.0).mean():.2f}%"
    ]
})
summary_stats.to_csv('earthquake_summary_statistics.csv', index=False)
print(f"\n✓ Summary statistics exported to: earthquake_summary_statistics.csv")
```

```python
# ============================================================================
# SECTION 4: SPATIAL-TEMPORAL GRIDDING AND FEATURE ENGINEERING
# ============================================================================

print("\n" + "=" * 80)
print("FEATURE ENGINEERING WITH PROPER TEMPORAL SEPARATION")
print("=" * 80 + "\n")

def create_spatial_grid(df, grid_size_km=25):
    """
    Create spatial grid cells (approximately grid_size_km x grid_size_km)
    1 degree latitude ≈ 111 km, 1 degree longitude ≈ 85 km at SF latitude

    This implements the 25km spatial tolerance specified in the proposal.
    """
    lat_step = grid_size_km / 111.0
    lon_step = grid_size_km / 85.0

    df['grid_lat'] = (df['latitude'] / lat_step).astype(int)
    df['grid_lon'] = (df['longitude'] / lon_step).astype(int)
    df['grid_id'] = df['grid_lat'].astype(str) + '_' + df['grid_lon'].astype(str)

    return df

earthquakes_df = create_spatial_grid(earthquakes_df, grid_size_km=25)
print(f"✓ Created {earthquakes_df['grid_id'].nunique()} spatial grid cells (25km × 25km)")

def create_temporal_windows(df, window_days=7):
    """
    Create temporal windows for forecasting
    7-day windows align with proposal's forecast horizon
    """
    df = df.sort_values('time').reset_index(drop=True)
    df['week_id'] = ((df['time'] - df['time'].min()).dt.days // window_days).astype(int)
    return df

earthquakes_df = create_temporal_windows(earthquakes_df, window_days=7)
print(f"✓ Created {earthquakes_df['week_id'].nunique()} temporal windows (7-day periods)")

def engineer_features(df, lookback_weeks=4, forecast_horizon_weeks=1):
    """
    Engineer temporal-spatial features with PROPER temporal separation to prevent data leakage.

    Key Design Principles:
    1. PAST DATA ONLY: Features use only historical data (weeks < current week)
    2. FUTURE TARGET: Target looks ahead (weeks > current week) with tolerance window
    3. NO LEAKAGE: Current week is NEVER used in features or overlapped with target
    4. REALISTIC FORECAST: Predicts 1-2 weeks ahead (operational horizon)

    Parameters:
    -----------
    df : DataFrame
        Earthquake data with grid_id and week_id
    lookback_weeks : int (default=4)
        Number of past weeks to use for feature calculation
    forecast_horizon_weeks : int (default=1)
        Number of future weeks to forecast (creates tolerance window)
        E.g., forecast_horizon_weeks=1 means predicting weeks [current+1, current+2]

    Returns:
    --------
    DataFrame with features and properly separated target variable
    """
    features_list = []

    print(f"Engineering features with {lookback_weeks}-week lookback and {forecast_horizon_weeks}-week forecast horizon...")

    # Get all unique grid-week combinations
    grid_weeks = df.groupby(['grid_id', 'week_id']).size().reset_index()[['grid_id', 'week_id']]

    for idx, row in grid_weeks.iterrows():
        grid_id = row['grid_id']
        week_id = row['week_id']

        # =====================================================================
        # HISTORICAL DATA (PAST ONLY - NO LEAKAGE)
        # =====================================================================
        # Use only data from BEFORE the current week (strict inequality)
        historical = df[(df['grid_id'] == grid_id) &
                       (df['week_id'] < week_id) &  # STRICTLY BEFORE current week
                       (df['week_id'] >= week_id - lookback_weeks)]

        # Need sufficient historical data to compute features
        if len(historical) < 5:  # Minimum threshold for reliable features
            continue

        # =====================================================================
        # FUTURE TARGET WITH TOLERANCE (PREVENTS LEAKAGE)
        # =====================================================================
        # Look AHEAD to future weeks with tolerance window
        # Example: If forecast_horizon_weeks=1, predicts weeks [week_id+1, week_id+2]
        # This gives 7-14 day forecast with ±7 day tolerance
        future_window = df[(df['grid_id'] == grid_id) &
                          (df['week_id'] > week_id) &  # STRICTLY AFTER current week
                          (df['week_id'] <= week_id + forecast_horizon_weeks + 1)]

        # Target: Is there at least one M≥3.0 event in the forecast window?
        target = (future_window['magnitude'] >= 3.0).any() if len(future_window) > 0 else False

        # Skip if we don't have future data to evaluate (at dataset boundaries)
        if len(future_window) == 0:
            continue

        # =====================================================================
        # FEATURE ENGINEERING (USING HISTORICAL DATA ONLY)
        # =====================================================================

        # Feature 1: Event count in lookback period
        event_count = len(historical)

        # Feature 2: Max magnitude in lookback
        max_mag = historical['magnitude'].max()

        # Feature 3: Mean magnitude in lookback
        mean_mag = historical['magnitude'].mean()

        # Feature 4: Cumulative seismic energy release (10^(1.5*M))
        # Uses Gutenberg-Richter energy-magnitude relationship
        energy = np.sum(10 ** (1.5 * historical['magnitude']))
        log_energy = np.log10(energy + 1)  # Log transform for scale

        # Feature 5: b-value estimation (Gutenberg-Richter law)
        # b-value indicates stress state: lower b → higher stress
        if len(historical) >= 10:
            mags = historical['magnitude'].values
            mag_bins = np.arange(mags.min(), mags.max() + 0.1, 0.1)
            counts, _ = np.histogram(mags, bins=mag_bins)

            # Compute b-value from log-linear fit
            if np.sum(counts > 0) > 1:
                valid_bins = mag_bins[:-1][counts > 0]
                log_counts = np.log10(counts[counts > 0] + 1)

                # b-value is negative slope of log(N) vs M
                try:
                    b_value = -np.polyfit(valid_bins, log_counts, 1)[0]
                    b_value = np.clip(b_value, 0.1, 3.0)  # Reasonable physical range
                except:
                    b_value = 1.0  # Default value
            else:
                b_value = 1.0
        else:
            b_value = 1.0

        # Feature 6: Seismicity rate change (acceleration/deceleration)
        # Compare recent activity to earlier baseline
        midpoint = week_id - lookback_weeks // 2
        recent = historical[historical['week_id'] >= midpoint]
        earlier = historical[historical['week_id'] < midpoint]

        recent_rate = len(recent) / (lookback_weeks // 2) if lookback_weeks > 0 else 0
        earlier_rate = len(earlier) / (lookback_weeks // 2) if lookback_weeks > 0 else 0
        rate_change = recent_rate - earlier_rate

        # Feature 7: Mean hypocentral depth
        mean_depth = historical['depth'].mean()

        # Feature 8: Spatial clustering coefficient
        # Lower std → more clustered → potential fault activation
        if len(historical) > 1:
            lat_std = historical['latitude'].std()
            lon_std = historical['longitude'].std()
            spatial_std = np.sqrt(lat_std**2 + lon_std**2)
        else:
            spatial_std = 0.0

        # Feature 9: Time since last event (recurrence interval)
        last_event_time = historical['time'].max()
        current_time = df[df['week_id'] == week_id]['time'].min()
        time_since_last = (current_time - last_event_time).days if pd.notna(current_time) else 365
        time_since_last = np.clip(time_since_last, 0, 365)  # Cap at 1 year

        # Feature 10: Magnitude variability (heterogeneity)
        mag_variance = historical['magnitude'].var() if len(historical) > 1 else 0.0

        # Feature 11: Recent swarm activity indicator
        # Count events in last week vs typical rate
        very_recent = historical[historical['week_id'] >= week_id - 1]
        swarm_indicator = len(very_recent) / (event_count / lookback_weeks) if event_count > 0 else 0

        # Feature 12: Average interevent time (regularity)
        if len(historical) > 1:
            historical_sorted = historical.sort_values('time')
            time_diffs = historical_sorted['time'].diff().dt.total_seconds() / 86400  # Convert to days
            avg_interevent_time = time_diffs.mean()
        else:
            avg_interevent_time = 365

        # Get timestamp for this observation (start of current week)
        timestamp = df[(df['grid_id'] == grid_id) & (df['week_id'] == week_id)]['time'].min()

        # =====================================================================
        # STORE FEATURES AND TARGET
        # =====================================================================
        features_list.append({
            'grid_id': grid_id,
            'week_id': week_id,
            'event_count': event_count,
            'max_magnitude': max_mag,
            'mean_magnitude': mean_mag,
            'cumulative_energy': log_energy,
            'b_value': b_value,
            'rate_change': rate_change,
            'mean_depth': mean_depth,
            'spatial_std': spatial_std,
            'time_since_last': time_since_last,
            'magnitude_variance': mag_variance,
            'swarm_indicator': swarm_indicator,
            'avg_interevent_time': avg_interevent_time,
            'target': int(target),
            'timestamp': timestamp,
            'forecast_start_week': week_id + 1,  # For tracking
            'forecast_end_week': week_id + forecast_horizon_weeks + 1
        })

    feature_df = pd.DataFrame(features_list)

    print(f"✓ Feature engineering complete")
    print(f"  - Total observations: {len(feature_df)}")
    print(f"  - Lookback window: {lookback_weeks} weeks (past data only)")
    print(f"  - Forecast horizon: {forecast_horizon_weeks + 1} weeks (future target)")
    print(f"  - Temporal separation: Current week NEVER overlaps with target")

    return feature_df

# Run feature engineering with proper temporal separation
print("\nEngineering features (this may take 1-2 minutes)...")
feature_df = engineer_features(earthquakes_df, lookback_weeks=4, forecast_horizon_weeks=1)

print(f"\n✓ Created feature matrix: {feature_df.shape}")
print(f"✓ Features: {feature_df.shape[1] - 5} predictive features")  # Minus metadata columns
print(f"✓ Target distribution: {feature_df['target'].sum()} positive samples ({100*feature_df['target'].mean():.2f}%)")

# Add feature: Distance to major faults (San Andreas, Hayward, Calaveras)
# Approximate fault locations in Bay Area
FAULTS = {
    'San Andreas': {'lat': 37.5, 'lon': -122.5},
    'Hayward': {'lat': 37.7, 'lon': -122.1},
    'Calaveras': {'lat': 37.4, 'lon': -121.8}
}

def distance_to_fault(lat, lon, fault_lat, fault_lon):
    """Calculate approximate distance in km using Haversine formula"""
    lat1, lon1 = np.radians(lat), np.radians(lon)
    lat2, lon2 = np.radians(fault_lat), np.radians(fault_lon)
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return 6371 * c  # Earth radius in km

# Add fault proximity features
for idx, row in feature_df.iterrows():
    grid_id = row['grid_id']
    grid_lat, grid_lon = map(int, grid_id.split('_'))
    # Convert grid indices back to lat/lon (approximate center)
    center_lat = grid_lat * (50 / 111.0)
    center_lon = grid_lon * (50 / 85.0)

    # Distance to nearest major fault
    min_fault_dist = min([
        distance_to_fault(center_lat, center_lon, fault['lat'], fault['lon'])
        for fault in FAULTS.values()
    ])
    feature_df.at[idx, 'fault_distance'] = min_fault_dist

print(f"\n✓ Added fault proximity features")
print(f"  Mean distance to nearest fault: {feature_df['fault_distance'].mean():.1f} km")

# Verify no data leakage
print("\n" + "=" * 80)
print("DATA LEAKAGE VERIFICATION")
print("=" * 80)
sample_row = feature_df.iloc[100]
print(f"\nExample observation:")
print(f"  Current week: {sample_row['week_id']}")
print(f"  Features computed from: weeks {sample_row['week_id']-4} to {sample_row['week_id']-1}")
print(f"  Target forecasts: weeks {sample_row['forecast_start_week']} to {sample_row['forecast_end_week']}")
print(f"  Temporal gap: ✓ No overlap (forecast is {sample_row['forecast_start_week'] - sample_row['week_id']} weeks ahead)")

# Display sample features
print("\n" + "=" * 80)
print("SAMPLE FEATURES")
print("=" * 80)
print(feature_df[['week_id', 'event_count', 'max_magnitude', 'b_value',
                  'rate_change', 'time_since_last', 'target']].head(10).to_string())
```

```python
# ============================================================================
# SECTION 6: BASELINE MODELS (WITH PROPER TOLERANCE MATCHING) - FIXED & OPTIMIZED
# ============================================================================

print("\n" + "=" * 80)
print("BASELINE MODELS")
print("=" * 80 + "\n")

print("Establishing baselines for comparison with ML models...")
print("These represent 'naive' forecasting strategies used in seismology.\n")

# ---------------------------------------------------------------------------
# Common settings
# ---------------------------------------------------------------------------
tolerance_weeks = 1  # must match tolerance used in target creation
lookback_weeks_for_persistence = 1  # how many prior weeks to inspect (can tune)

# Ensure index alignment for test_df iteration
test_idx_list = test_df.index.tolist()

# ---------------------------------------------------------------------------
# Baseline 1: Uniform Rate Model (Null Model)
# ---------------------------------------------------------------------------
uniform_rate = y_train.mean()
uniform_preds = np.full(len(y_test), uniform_rate)
uniform_brier = brier_score_loss(y_test, uniform_preds)

print(f"Baseline 1 - Uniform Rate Model (Null Hypothesis):")
print(f"  Strategy: Constant probability = historical average")
print(f"  Historical rate: {uniform_rate:.4f} ({100*uniform_rate:.2f}% of time periods)")
print(f"  Brier score: {uniform_brier:.4f}")
print(f"  Interpretation: Assumes no predictive information available\n")

# ---------------------------------------------------------------------------
# Baseline 2: Temporal Persistence with Tolerance Window
# ---------------------------------------------------------------------------
print(f"Baseline 2 - Temporal Persistence (with ±{tolerance_weeks} week tolerance):")

persistence_preds = []
persistence_lookback_counts = []

# Precompute dictionary mapping (grid_id, week_id) -> target (train)
train_target_map = {(r['grid_id'], int(r['week_id'])): int(r['target']) for _, r in train_df.iterrows()}

for idx in test_idx_list:
    grid = test_df.loc[idx, 'grid_id']
    week = int(test_df.loc[idx, 'week_id'])

    # consider prior weeks up to lookback_weeks_for_persistence + tolerance
    prior_weeks = list(range(week - tolerance_weeks - lookback_weeks_for_persistence,
                             week))
    recent_vals = [train_target_map.get((grid, w), None) for w in prior_weeks]
    recent_vals = [v for v in recent_vals if v is not None]

    if len(recent_vals) > 0:
        # probabilistic: fraction of recent windows that were active (gives a probability)
        prob = np.mean(recent_vals)
        persistence_preds.append(prob)
        persistence_lookback_counts.append(len(recent_vals))
    else:
        # fallback to climatology
        persistence_preds.append(uniform_rate)
        persistence_lookback_counts.append(0)

persistence_preds = np.array(persistence_preds)
persistence_brier = brier_score_loss(y_test, persistence_preds)

persistence_binary = (persistence_preds > 0.5).astype(int)
persistence_active_rate = np.mean(persistence_binary)
persistence_with_data = np.sum(np.array(persistence_lookback_counts) > 0) / len(persistence_lookback_counts)

print(f"  Brier score: {persistence_brier:.4f}")
print(f"  Forecasts 'active' (>0.5): {100*persistence_active_rate:.2f}%")
print(f"  Grid-weeks with recent data used: {100*persistence_with_data:.2f}%")
print(f"  Interpretation: captures short-term temporal clustering (aftershocks)\n")

# ---------------------------------------------------------------------------
# Baseline 3: Spatial Persistence (Neighboring Grid Cells)
# ---------------------------------------------------------------------------
print("Baseline 3 - Spatial Persistence (neighbor grids):")

spatial_persistence_preds = []

# Precompute a mapping from grid_id -> (grid_lat, grid_lon)
# If your grid ids are strings like "123_456", convert accordingly
def parse_grid(grid_id):
    lat_s, lon_s = grid_id.split('_')
    return int(lat_s), int(lon_s)

# Create a lookup of train targets grouped by (grid_id, week_id) as before
# We'll reuse train_target_map

for idx in test_idx_list:
    grid = test_df.loc[idx, 'grid_id']
    week = int(test_df.loc[idx, 'week_id'])
    lat_i, lon_i = parse_grid(grid)

    # gather neighbor grid ids (8-connectivity + self)
    neighbor_grids = [f"{lat_i + dlat}_{lon_i + dlon}"
                      for dlat in (-1, 0, 1) for dlon in (-1, 0, 1)]

    # check prior weeks in neighbors
    prior_weeks = list(range(week - tolerance_weeks - lookback_weeks_for_persistence, week))
    neighbor_vals = []
    for ng in neighbor_grids:
        neighbor_vals += [train_target_map.get((ng, w), None) for w in prior_weeks]
    neighbor_vals = [v for v in neighbor_vals if v is not None]

    if len(neighbor_vals) > 0:
        # probability = fraction of neighbor grid-weeks that were active
        spatial_persistence_preds.append(np.mean(neighbor_vals))
    else:
        spatial_persistence_preds.append(uniform_rate)

spatial_persistence_preds = np.array(spatial_persistence_preds)
spatial_brier = brier_score_loss(y_test, spatial_persistence_preds)
spatial_active_rate = np.mean((spatial_persistence_preds > 0.5).astype(int))

print(f"  Brier score: {spatial_brier:.4f}")
print(f"  Forecasts 'active' (>0.5): {100*spatial_active_rate:.2f}%")
print(f"  Interpretation: accounts for activity spilling across grid boundaries\n")

# ---------------------------------------------------------------------------
# Baseline 4: Gutenberg-Richter recurrence per grid
# ---------------------------------------------------------------------------
print("Baseline 4 - Gutenberg-Richter recurrence (per-grid magnitude-frequency):")

# We'll cache per-grid parameters to avoid repeated computation
from collections import defaultdict
gr_rate_map = {}  # grid -> predicted probability for M>=3.0 (fallback to uniform_rate if insufficient data)

# Precompute events per grid from the raw earthquakes_df (not aggregated features),
# but only using events that occur before each test week (we'll do an approximate per-grid rate using train_df counts)
# Simpler approach: use train_df's historical counts and b_value column if present; otherwise compute from earthquakes_df.

# If your 'train_df' includes b_value or counts, use them directly:
if 'b_value' in train_df.columns:
    # compute grid-level avg b and avg rate per week from train_df
    grid_week_counts = train_df.groupby('grid_id')['target'].agg(['sum','count']).rename(columns={'sum':'n_significant','count':'n_weeks'}).reset_index()
    grid_week_counts['rate_per_week'] = grid_week_counts['n_significant'] / grid_week_counts['n_weeks']
    grid_rate_dict = grid_week_counts.set_index('grid_id')['rate_per_week'].to_dict()
else:
    # fallback: compute rate per grid from earthquakes_df using train period events
    train_time_cutoff = train_df['timestamp'].max()
    eq_train = earthquakes_df[earthquakes_df['time'] <= train_time_cutoff]
    # count weeks per grid in train_df to normalize
    weeks_per_grid = train_df.groupby('grid_id')['week_id'].nunique().to_dict()
    sig_counts = eq_train[eq_train['magnitude'] >= 3.0].groupby('grid_id').size().to_dict()
    grid_rate_dict = {}
    for g, wcount in weeks_per_grid.items():
        n_sig = sig_counts.get(g, 0)
        grid_rate_dict[g] = n_sig / wcount if wcount > 0 else uniform_rate

# For each test sample, compute a GR-derived probability using the per-grid historical rate (as a proxy).
gr_preds = []
for idx in test_idx_list:
    grid = test_df.loc[idx, 'grid_id']
    rate = grid_rate_dict.get(grid, uniform_rate)
    # Use rate as probability proxy — this is a pragmatic, robust approach:
    # If you want a pure GR formula you'd need robust b-value fits per grid (often not possible with low counts).
    gr_preds.append(min(max(rate, 0.0), 1.0))

gr_preds = np.array(gr_preds)
gr_brier = brier_score_loss(y_test, gr_preds)
gr_active_rate = np.mean((gr_preds > 0.5).astype(int))

print(f"  Brier score: {gr_brier:.4f}")
print(f"  Forecasts 'active' (>0.5): {100*gr_active_rate:.2f}%")
print(f"  Interpretation: uses historical magnitude-frequency as a probabilistic climatology\n")

# ---------------------------------------------------------------------------
# Baseline 5: Simplified ETAS-Inspired Model (efficient approximation)
# ---------------------------------------------------------------------------
print("Baseline 5 - ETAS-Inspired simplified model (limited lookback, efficient):")

# ETAS parameters (coarse values)
mu = uniform_rate  # background rate (weekly)
K = 0.15           # productivity scale (tunable)
alpha = 1.2        # magnitude scaling
c = 0.1            # time offset (weeks)
p = 1.1            # Omori exponent
max_lookback_weeks = 52  # limit to one year to keep computation bounded

# Pre-build a per-grid list of events (time, magnitude) from earthquakes_df within train period for faster access
train_time_cutoff = train_df['timestamp'].max()
eq_train = earthquakes_df[earthquakes_df['time'] <= train_time_cutoff].copy()

grid_event_map = defaultdict(list)
for _, e in eq_train.iterrows():
    grid_event_map[e['grid_id']].append((e['time'], e['magnitude']))

# Now compute ETAS-inspired probability per test row
etas_preds = []
for idx in test_idx_list:
    grid = test_df.loc[idx, 'grid_id']
    current_time = test_df.loc[idx, 'timestamp']

    events = grid_event_map.get(grid, [])
    if len(events) == 0:
        etas_preds.append(mu)
        continue

    triggered_rate = 0.0
    # iterate events (small per-grid counts since dataset isn't huge)
    for ev_time, ev_mag in events:
        # weeks since event
        time_since_weeks = (current_time - ev_time).days / 7.0
        if time_since_weeks <= 0 or time_since_weeks > max_lookback_weeks:
            continue
        decay = 1.0 / ((time_since_weeks + c) ** p)
        productivity = K * np.exp(alpha * (ev_mag - 1.5))
        triggered_rate += productivity * decay

    total_rate = mu + triggered_rate
    # map rate to probability: if rate is small (<<1) it's already interpretable as weekly probability; clip to [0,1]
    etas_preds.append(min(max(total_rate, 0.0), 1.0))

etas_preds = np.array(etas_preds)
etas_brier = brier_score_loss(y_test, etas_preds)
etas_active_rate = np.mean((etas_preds > 0.5).astype(int))

print(f"  Brier score: {etas_brier:.4f}")
print(f"  Forecasts 'active' (>0.5): {100*etas_active_rate:.2f}%")
print(f"  Interpretation: physics-inspired clustering (approximate ETAS)\n")

# ---------------------------------------------------------------------------
# Baseline summary table & storing results
# ---------------------------------------------------------------------------
baseline_results = {
    'Uniform Rate': {'brier': uniform_brier, 'preds': uniform_preds},
    'Temporal Persistence': {'brier': persistence_brier, 'preds': persistence_preds},
    'Spatial Persistence': {'brier': spatial_brier, 'preds': spatial_persistence_preds},
    'Gutenberg-Richter': {'brier': gr_brier, 'preds': gr_preds},
    'ETAS-Inspired': {'brier': etas_brier, 'preds': etas_preds}
}

print("\n" + "-" * 80)
print("BASELINE SUMMARY - Statistical + Seismological Models")
print("-" * 80)
print(f"{'Model':<30} {'Brier Score':<15} {'Active Rate':<15} {'Type':<20}")
print("-" * 80)
print(f"{'Uniform Rate (Null)':<30} {uniform_brier:<15.4f} {100*uniform_rate:>13.2f}%  {'Statistical':<20}")
print(f"{'Temporal Persistence':<30} {persistence_brier:<15.4f} {100*persistence_active_rate:>13.2f}%  {'Statistical':<20}")
print(f"{'Spatial Persistence':<30} {spatial_brier:<15.4f} {100*spatial_active_rate:>13.2f}%  {'Statistical':<20}")
print(f"{'Gutenberg-Richter':<30} {gr_brier:<15.4f} {100*gr_active_rate:>13.2f}%  {'Seismological':<20}")
print(f"{'ETAS-Inspired':<30} {etas_brier:<15.4f} {100*etas_active_rate:>13.2f}%  {'Seismological':<20}")
print("-" * 80)

# Identify best baseline
best_baseline_name = min(baseline_results.items(), key=lambda kv: kv[1]['brier'])[0]
best_baseline = baseline_results[best_baseline_name]['brier']
print(f"\nBest baseline: {best_baseline_name} with Brier score {best_baseline:.4f}")
print(f"ML models must beat this to demonstrate forecasting skill!")
print("\nKey Insights:")
print(f"  - G-R model captures magnitude statistics: {'Improved' if gr_brier < uniform_brier else 'Limited'}")
print(f"  - ETAS captures temporal clustering: {'Improved' if etas_brier < spatial_brier else 'Limited'}")
print(f"  - Best seismological baseline: {best_baseline_name}\n")
```

```python
# ============================================================================
# SECTION 7: MACHINE LEARNING MODELS (11 MODELS)
# ============================================================================

print("\n" + "=" * 80)
print("MACHINE LEARNING MODELS - COMPREHENSIVE COMPARISON")
print("=" * 80 + "\n")

print("Note: ARIMA and Hidden Markov Models (mentioned in proposal) are designed for")
print("univariate time series only. Our spatial-temporal grid dataset is multivariate")
print("and nonlinear, so we use neural networks + LSTM-approximation instead.\n")

from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import brier_score_loss, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
import lightgbm as lgb
import xgboost as xgb

models_results = {}

# --------------------------
# Utility: evaluate model
# --------------------------
def evaluate_model(name, model, X_test, y_test, calibrated=False):
    if calibrated:
        model = CalibratedClassifierCV(model, method='sigmoid', cv=3)
        model.fit(X_train_scaled, y_train)

    probs = model.predict_proba(X_test)[:, 1]
    brier = brier_score_loss(y_test, probs)
    auc = roc_auc_score(y_test, probs)

    models_results[name] = {
        'model': model,
        'probs': probs,
        'brier': brier,
        'auc': auc
    }

    print(f"  Brier: {brier:.4f}, AUC: {auc:.4f}")


# --------------------------
# 1. Logistic Regression
# --------------------------
print("1/11 Training Logistic Regression...")
lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
lr_cal = CalibratedClassifierCV(lr, method='sigmoid', cv=3)
lr_cal.fit(X_train_scaled, y_train)
evaluate_model("Logistic Regression", lr_cal, X_test_scaled, y_test)


# --------------------------
# 2. Random Forest
# --------------------------
print("2/11 Training Random Forest...")
rf = RandomForestClassifier(
    n_estimators=150,
    max_depth=12,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)
rf_cal = CalibratedClassifierCV(rf, method='sigmoid', cv=3)
rf_cal.fit(X_train_scaled, y_train)
evaluate_model("Random Forest", rf_cal, X_test_scaled, y_test)


# --------------------------
# 3. Gradient Boosting
# --------------------------
print("3/11 Training Gradient Boosting...")
gb = GradientBoostingClassifier(
    n_estimators=150,
    learning_rate=0.08,
    max_depth=4,
    random_state=42
)
gb.fit(X_train_scaled, y_train)
evaluate_model("Gradient Boosting", gb, X_test_scaled, y_test)


# --------------------------
# 4. XGBoost
# --------------------------
print("4/11 Training XGBoost...")
scale_pos_weight = (y_train == 0).sum() / max(1, (y_train == 1).sum())

xgb_model = xgb.XGBClassifier(
    n_estimators=150,
    max_depth=4,
    learning_rate=0.1,
    objective='binary:logistic',
    scale_pos_weight=scale_pos_weight,
    eval_metric='logloss',
    random_state=42
)
xgb_model.fit(X_train_scaled, y_train)
evaluate_model("XGBoost", xgb_model, X_test_scaled, y_test)


# --------------------------
# 5. LightGBM
# --------------------------
print("5/11 Training LightGBM...")
lgb_model = lgb.LGBMClassifier(
    n_estimators=150,
    max_depth=4,
    learning_rate=0.1,
    class_weight='balanced',
    random_state=42
)
lgb_model.fit(X_train_scaled, y_train)
evaluate_model("LightGBM", lgb_model, X_test_scaled, y_test)


# --------------------------
# 6. SVM (RBF)
# --------------------------
print("6/11 Training Support Vector Machine (RBF)...")
svm = SVC(kernel='rbf', C=1.0, class_weight='balanced', probability=True, random_state=42)
svm.fit(X_train_scaled, y_train)
evaluate_model("SVM (RBF)", svm, X_test_scaled, y_test)


# --------------------------
# 7. KNN
# --------------------------
print("7/11 Training K-Nearest Neighbors...")
knn = KNeighborsClassifier(n_neighbors=15, weights='distance')
knn.fit(X_train_scaled, y_train)
evaluate_model("K-Nearest Neighbors", knn, X_test_scaled, y_test)


# --------------------------
# 8. Neural Network (MLP)
# --------------------------
print("8/11 Training Neural Network (MLP)...")
mlp = MLPClassifier(
    hidden_layer_sizes=(50, 25),
    activation='relu',
    max_iter=700,
    random_state=42,
    early_stopping=True
)
mlp.fit(X_train_scaled, y_train)
evaluate_model("Neural Network (MLP)", mlp, X_test_scaled, y_test)


# --------------------------
# 9. Naive Bayes
# --------------------------
print("9/11 Training Naive Bayes...")
nb = GaussianNB()
nb.fit(X_train_scaled, y_train)
evaluate_model("Naive Bayes", nb, X_test_scaled, y_test)


# --------------------------
# 10. Decision Tree
# --------------------------
print("10/11 Training Decision Tree...")
dt = DecisionTreeClassifier(max_depth=10, class_weight='balanced', random_state=42)
dt_cal = CalibratedClassifierCV(dt, method='sigmoid', cv=3)
dt_cal.fit(X_train_scaled, y_train)
evaluate_model("Decision Tree", dt_cal, X_test_scaled, y_test)


# --------------------------
# 11. LSTM Approximation
# --------------------------
print("11/11 Training LSTM-Approx Model...")
print("  (Adds seasonality + long-term history as features)")

# add temporal features
X_train_temporal = np.column_stack([
    X_train_scaled,
    train_df['week_id'] % 52,
    (train_df['week_id'] // 52)
])
X_test_temporal = np.column_stack([
    X_test_scaled,
    test_df['week_id'] % 52,
    (test_df['week_id'] // 52)
])

lstm_approx = MLPClassifier(
    hidden_layer_sizes=(100, 50, 25),
    activation='tanh',
    max_iter=800,
    random_state=42,
    early_stopping=True
)
lstm_approx.fit(X_train_temporal, y_train)
evaluate_model("LSTM-Approx", lstm_approx, X_test_temporal, y_test)


# ============================================================================
# EXPORT RESULTS
# ============================================================================
model_comparison = pd.DataFrame([
    {'Model': name, 'Brier Score': r['brier'], 'ROC-AUC': r['auc']}
    for name, r in models_results.items()
]).sort_values('Brier Score')

model_comparison.to_csv("model_comparison_results.csv", index=False)

print("\n✓ Model comparison exported → model_comparison_results.csv\n")

print("=" * 80)
print("MODEL RANKING BY BRIER SCORE (Lower = Better)")
print("=" * 80)
for i, row in model_comparison.iterrows():
    print(f"{row['Model']:<25}  Brier: {row['Brier Score']:.4f} | AUC: {row['ROC-AUC']:.4f}")
```

```python
# ============================================================================
# SECTION 8: MODEL EVALUATION AND COMPARISON (WITH THRESHOLD ANALYSIS)
# ============================================================================

print("\n" + "=" * 80)
print("MODEL COMPARISON AND EVALUATION")
print("=" * 80 + "\n")

# ============================================================================
# Part 1: Brier Skill Scores (Model Performance vs Baselines)
# ============================================================================
print("=" * 80)
print("BRIER SKILL SCORES (Relative to Best Baseline)")
print("=" * 80)

# Use best baseline as reference
best_baseline_brier = min(baseline_results['Uniform Rate']['brier'],
                          baseline_results['Temporal Persistence']['brier'],
                          baseline_results['Spatial Persistence']['brier'])

print(f"\nReference baseline: {best_baseline_brier:.4f} (best of three baselines)")
print(f"\nBaseline comparison:")
for baseline_name, baseline_data in baseline_results.items():
    bss = 1 - baseline_data['brier'] / best_baseline_brier
    print(f"  {baseline_name:<25} Brier: {baseline_data['brier']:.4f}  BSS: {bss:>6.3f}")

print(f"\nMachine Learning models:")
for name, results in models_results.items():
    bss = 1 - results['brier'] / best_baseline_brier
    improvement = (best_baseline_brier - results['brier']) / best_baseline_brier * 100
    print(f"  {name:<25} Brier: {results['brier']:.4f}  BSS: {bss:>6.3f}  ({improvement:+.1f}% improvement)")

print("\nInterpretation:")
print("  BSS > 0: Model has skill (better than baseline)")
print("  BSS = 0: Model equals baseline (no added value)")
print("  BSS < 0: Model worse than baseline (counterproductive)")

# ============================================================================
# Part 2: Threshold Sensitivity Analysis
# ============================================================================
print("\n" + "=" * 80)
print("OPERATIONAL THRESHOLD ANALYSIS")
print("=" * 80)
print("\nEvaluating models at multiple decision thresholds...")
print("This reveals the precision-recall tradeoff for operational deployment.\n")
print("Key Metrics (per revised proposal):")
print("  - Hit Rate: % of actual M≥3.0 events successfully forecasted")
print("  - False Alarm Ratio (FAR): % of alarms that were false (target: ≤20%)")
print("  - Miss Rate: % of actual M≥3.0 events that were missed")
print("  - Success Rate: Overall accuracy within spatial/temporal tolerances")
print("  - Precision: 1 - FAR (% of alarms that were correct)")
print("\nSpatial/Temporal Tolerance (per proposal):")
print("  - Spatial: 25km grid cells (±12.5km from grid center)")
print("  - Temporal: ±1 week forecast horizon (weeks ahead)")
print("  A forecast is successful if M≥3.0 occurs within these tolerances\n")

thresholds = [0.3, 0.4, 0.5, 0.6, 0.7]
threshold_results = {}

for thresh in thresholds:
    print("=" * 80)
    print(f"THRESHOLD = {thresh:.1%} (Forecast 'High Risk' if probability ≥ {thresh:.1%})")
    print("=" * 80)

    threshold_results[thresh] = {}

    for name, results in models_results.items():
        probs = results['probs']
        preds = (probs >= thresh).astype(int)

        # Confusion matrix elements
        tp = np.sum((preds == 1) & (y_test == 1))
        fp = np.sum((preds == 1) & (y_test == 0))
        fn = np.sum((preds == 0) & (y_test == 1))
        tn = np.sum((preds == 0) & (y_test == 0))

        # Operational metrics
        hit_rate = tp / (tp + fn) if (tp + fn) > 0 else 0  # Recall/Sensitivity
        false_alarm_ratio = fp / (fp + tp) if (fp + tp) > 0 else 0  # FAR = FP / (FP + TP)
        alarm_rate = np.mean(preds)  # Overall alarm frequency
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0  # 1 - FAR
        f1_score = 2 * precision * hit_rate / (precision + hit_rate) if (precision + hit_rate) > 0 else 0
        miss_rate = fn / (tp + fn) if (tp + fn) > 0 else 0  # Complement of hit rate
        success_rate = (tp + tn) / (tp + tn + fp + fn) if (tp + tn + fp + fn) > 0 else 0  # Overall accuracy

        # Store for visualization
        threshold_results[thresh][name] = {
            'hit_rate': hit_rate,
            'false_alarm_ratio': false_alarm_ratio,
            'alarm_rate': alarm_rate,
            'precision': precision,
            'f1': f1_score,
            'miss_rate': miss_rate,
            'success_rate': success_rate,
            'tp': tp, 'fp': fp, 'fn': fn, 'tn': tn
        }

        # Format output - Check against 20% FAR constraint from proposal
        acceptable_far = "✓" if false_alarm_ratio <= 0.20 else "✗"

        print(f"\n{name}:")
        print(f"  Hit Rate (Recall):      {hit_rate:>6.1%}  (captured {tp}/{tp+fn} significant events)")
        print(f"  False Alarm Ratio:      {false_alarm_ratio:>6.1%}  {acceptable_far} (target: ≤20%, per proposal)")
        print(f"  Miss Rate:              {miss_rate:>6.1%}  (missed {fn}/{tp+fn} events)")
        print(f"  Success Rate:           {success_rate:>6.1%}  (overall accuracy with tolerances)")
        print(f"  Precision:              {precision:>6.1%}  (when alarm raised, {precision:.1%} correct)")
        print(f"  F1 Score:               {f1_score:>6.3f}")

# ============================================================================
# Part 3: Optimal Threshold Selection
# ============================================================================
print("\n" + "=" * 80)
print("THRESHOLD SELECTION ANALYSIS")
print("=" * 80)

# Find optimal threshold for each model based on different criteria
print("\nOptimal thresholds by optimization criterion:\n")

for name, results in models_results.items():
    probs = results['probs']

    # Scan thresholds
    best_f1_thresh = 0
    best_f1 = 0
    best_constrained_thresh = 0
    best_constrained_f1 = 0

    for thresh in np.arange(0.1, 0.9, 0.05):
        preds = (probs >= thresh).astype(int)
        tp = np.sum((preds == 1) & (y_test == 1))
        fp = np.sum((preds == 1) & (y_test == 0))
        fn = np.sum((preds == 0) & (y_test == 1))

        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0
        alarm_rate = np.mean(preds)

        # Best F1 overall
        if f1 > best_f1:
            best_f1 = f1
            best_f1_thresh = thresh

        # Best F1 with alarm rate constraint
        if f1 > best_constrained_f1 and alarm_rate <= 0.20:
            best_constrained_f1 = f1
            best_constrained_thresh = thresh

    print(f"{name}:")
    print(f"  Max F1:                 {best_f1:.3f} at threshold {best_f1_thresh:.2f}")
    print(f"  Max F1 (alarm ≤20%):    {best_constrained_f1:.3f} at threshold {best_constrained_thresh:.2f}")
    print(f"  Proposal requirement:   threshold = 0.60 (as specified)")
    print()

# ============================================================================
# Part 4: Threshold Sweep Visualization Data
# ============================================================================
print("\n" + "=" * 80)
print("PREPARING THRESHOLD SWEEP VISUALIZATION")
print("=" * 80)

# Calculate metrics across threshold range for plotting
threshold_sweep = np.arange(0.1, 0.95, 0.05)
sweep_data = {name: {'hit_rate': [], 'precision': [], 'alarm_rate': [], 'f1': []}
              for name in models_results.keys()}

for thresh in threshold_sweep:
    for name, results in models_results.items():
        probs = results['probs']
        preds = (probs >= thresh).astype(int)

        tp = np.sum((preds == 1) & (y_test == 1))
        fp = np.sum((preds == 1) & (y_test == 0))
        fn = np.sum((preds == 0) & (y_test == 1))

        hit_rate = tp / (tp + fn) if (tp + fn) > 0 else 0
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        alarm_rate = np.mean(preds)
        f1 = 2 * precision * hit_rate / (precision + hit_rate) if (precision + hit_rate) > 0 else 0

        sweep_data[name]['hit_rate'].append(hit_rate)
        sweep_data[name]['precision'].append(precision)
        sweep_data[name]['alarm_rate'].append(alarm_rate)
        sweep_data[name]['f1'].append(f1)

# Create threshold sweep visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Hit Rate vs Threshold
ax1 = axes[0, 0]
for name in models_results.keys():
    ax1.plot(threshold_sweep, sweep_data[name]['hit_rate'],
             marker='o', markersize=3, linewidth=2, label=name)
ax1.axvline(0.60, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Proposal threshold')
ax1.set_xlabel('Probability Threshold')
ax1.set_ylabel('Hit Rate (Recall)')
ax1.set_title('Hit Rate vs Threshold\n(Higher is better, but raises false alarms)')
ax1.legend(loc='upper right')
ax1.grid(True, alpha=0.3)
ax1.set_xlim([0.1, 0.9])
ax1.set_ylim([0, 1])

# Plot 2: Precision vs Threshold
ax2 = axes[0, 1]
for name in models_results.keys():
    ax2.plot(threshold_sweep, sweep_data[name]['precision'],
             marker='o', markersize=3, linewidth=2, label=name)
ax2.axvline(0.60, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Proposal threshold')
ax2.set_xlabel('Probability Threshold')
ax2.set_ylabel('Precision')
ax2.set_title('Precision vs Threshold\n(Higher is better, but misses events)')
ax2.legend(loc='upper left')
ax2.grid(True, alpha=0.3)
ax2.set_xlim([0.1, 0.9])
ax2.set_ylim([0, 1])

# Plot 3: Alarm Rate vs Threshold
ax3 = axes[1, 0]
for name in models_results.keys():
    ax3.plot(threshold_sweep, sweep_data[name]['alarm_rate'],
             marker='o', markersize=3, linewidth=2, label=name)
ax3.axhline(0.20, color='orange', linestyle='--', linewidth=2, alpha=0.7, label='20% alarm limit')
ax3.axvline(0.60, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Proposal threshold')
ax3.set_xlabel('Probability Threshold')
ax3.set_ylabel('Alarm Rate')
ax3.set_title('Alarm Rate vs Threshold\n(Must stay below 20% for operational use)')
ax3.legend(loc='upper right')
ax3.grid(True, alpha=0.3)
ax3.set_xlim([0.1, 0.9])
ax3.set_ylim([0, 0.5])

# Plot 4: F1 Score vs Threshold
ax4 = axes[1, 1]
for name in models_results.keys():
    ax4.plot(threshold_sweep, sweep_data[name]['f1'],
             marker='o', markersize=3, linewidth=2, label=name)
ax4.axvline(0.60, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Proposal threshold')
ax4.set_xlabel('Probability Threshold')
ax4.set_ylabel('F1 Score')
ax4.set_title('F1 Score vs Threshold\n(Balances precision and recall)')
ax4.legend(loc='upper right')
ax4.grid(True, alpha=0.3)
ax4.set_xlim([0.1, 0.9])
ax4.set_ylim([0, 0.6])

plt.tight_layout()
plt.savefig('threshold_sensitivity_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Threshold sweep visualization saved as 'threshold_sensitivity_analysis.png'")

# ============================================================================
# Part 5: Summary Table
# ============================================================================
print("\n" + "=" * 80)
print("SUMMARY: BEST MODEL AT PROPOSAL THRESHOLD (60%)")
print("=" * 80)

best_model_name = min(models_results.items(), key=lambda x: x[1]['brier'])[0]
print(f"\nBest performing model: {best_model_name}")
print(f"Selected threshold: 60% (as specified in proposal)")
print(f"\nPerformance at 60% threshold:")

if 0.6 in threshold_results:
    best_metrics = threshold_results[0.6][best_model_name]
    print(f"  Hit Rate:           {best_metrics['hit_rate']:.2%}")
    print(f"  Precision:          {best_metrics['precision']:.2%}")
    print(f"  False Alarm Ratio:  {best_metrics['false_alarm_ratio']:.2%} {'✓ Within limit' if best_metrics['false_alarm_ratio'] <= 0.20 else '✗ Exceeds limit'}")
    print(f"  Miss Rate:          {best_metrics['miss_rate']:.2%}")
    print(f"  Alarm Rate:         {best_metrics['alarm_rate']:.2%}")
    print(f"  F1 Score:           {best_metrics['f1']:.3f}")
    print(f"  Events captured:    {best_metrics['tp']}/{best_metrics['tp'] + best_metrics['fn']}")
    print(f"  False alarms:       {best_metrics['fp']} (out of {best_metrics['tp'] + best_metrics['fp']} total alarms)")

print("\n" + "=" * 80)
```
